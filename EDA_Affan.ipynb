```python
sns.stripplot(x='DOC_DATE_YEAR', y='ORDER_STATUS', data=df, size=4, jitter=True)
plt.ylabel('ORDER_STATUS')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='ORDER_STATUS', y='DOC_DATE_WEEK_YR', data=df)
plt.ylabel('DOC_DATE_WEEK_YR')
plt.subplot(1,2,2)
sns.violinplot(x='ORDER_STATUS', y='DOC_DATE_WEEK_YR', data=df)
plt.ylabel('DOC_DATE_WEEK_YR')
plt.tight_layout()
plt.show()
```

```python
sns.violinplot(x='ORDER_STATUS', y='DOC_DATE_WEEK_YR', data=df, inner=None, color='lightgray')
sns.stripplot(x='ORDER_STATUS', y='DOC_DATE_WEEK_YR', data=df, size=4, jitter=True)
plt.ylabel('ORDER_STATUS')
plt.show()
```

```python
sns.jointplot(x= 'DOC_DATE_WEEK_YR', y= 'DOC_DATE_YEAR', data=df)
plt.show()
```
