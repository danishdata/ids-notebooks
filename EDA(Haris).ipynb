```python
(data['DOC_DATE_YEAR']
 .value_counts(sort=False)
 .plot.barh()
 .set_title('YEAR')
)

plt.show()
```

```python
data_num = data.select_dtypes(include = ['float64', 'int64'])
data_num.head()
```

```python
data_num['POP_CODE'] = pd.factorize(data_num.POP_CODE)[0] + 1
print (data_num)
```

```python
data_num.POP_CODE.unique()
```

```python
data_num.hist(figsize=(20, 24), bins=80, xlabelsize=16, ylabelsize=16)
plt.show()
```

```python
data_num.corr()
```

```python
import seaborn as sns

f, ax = plt.subplots(figsize=(10, 8))
corr = data_num.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
plt.show()
```

```python
data_num_corr = data_num.corr()['POP_CODE'][:-1] # -1 because the latest row is SalePrice
golden_features_list = data_num_corr[abs(data_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with POP_CODE:\n{}".format(len(golden_features_list), golden_features_list))
```
